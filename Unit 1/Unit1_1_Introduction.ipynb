## **Unit 1 -  Into the Nerd Herd: Revamping the Cardiovascular Department using Machine Learning** 

Cardiovascular disease is the leading cause of mortality, with an estimated **17 million** deaths each year. Every hospital has been encouraged to put in place initiatives:

-  to improve prevention and local population management 

- to help keep population healthy and 

- lessen the need for acute care, in order to reduce the strain on staff and beds available. 

AI can help enable earlier prevention, in particular the use of an automated echocardiogram service to support physicians, run screening test to detect cardiovascular disease, increasing accuracy, reducing variability and increasing confidence for less expert staff.

A group of hospital cardiologists are working together to define a screening methodology for GPs:
-  to identify early onset of cardiovascular disease in the community, 
- increase prevention and optimise  referral and 
- triage onto the appropriate hospital cardiology service. 

The aim is to:
-  focus healthcare provision on patients at high risk of cardiovascular disease, 
- improve the use of limited hospital resources, and - optimise scheduling and waiting times for patients. 

This is a GP service with the hospital supervision. 

The GP will use an AI software solution for risk prediction that triangulates the patient medical history together with all investigation results and any additional information reported (e.g. personal information, blood pressure, smoking, physical activity and any other relevant information for cardiovascular disease). This will help to identify patients that need to be referred to the hospital cardiology service for further examination. However, there have been some concerns that the AI software will effectively be screening asymptomatic individuals. With a tool that is not 100% specific or sensitive there’s a risk of false positives and false negatives. 

Patient, 45 male, arrives at GP following an occurrence of sudden sharp pain in the chest when exercising, the pain quickly receded when they stopped and rested. After some routine checks, the GP refers the patient to the Rapid Access Chest Pain Clinic for further investigation.

**Cardiology expert:** Hello, I am a cardiology consultant and lead the Cardiac Physiology Clinical Unit of a leading teaching hospital in Europe. The unit is very busy providing a wide range of patient procedures including 
- echocardiography,
- electrocardiography, 
- ambulatory blood pressure monitoring, and 
- nuclear cardiology imaging. 

There are a number of challenges facing the clinic. One of the most heavily used areas is echocardiology and so any improvements here will have significant impact for patients and the hospital. This area has long waiting times and delays to receiving prognosis after a scan. Patients are increasingly dissatisfied and staff are feeling increasing pressured and stressed. The service is seeing a slight increase in misdiagnosis and non-diagnostic quality images which is further worsening the situation. In addition, we’re seeing a trend of increasing age of the demographic of the community giving a slowly increasing demand for echocardiology services. 

I would like to introduce some new tools and technologies as a way of improving the unit and the patient care. I have read some journal articles showing Artificial Intelligence in the domain of imaging has demonstrated promising progress recently and could help address some of the unmet need in the department. Our overall aim is to identify new tools that will improve patient and staff experience, as well as increase efficiency and reduce waiting time.

However, AI is still a very new technology in healthcare and I am unfamiliar with some of the basics. I feel there is cultural resistance from my colleagues who are highly suspicious of AI despite having little exposure or experience.
Looking at some of the recent patient feedback, resource constraints and waiting times I organize a meeting with my colleagues to explore alternative options to improve the clinic.
