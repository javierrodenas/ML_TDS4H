### **What is Artificial Intelligence?**

Intuitively, AI refers to the ability of certain machines to behave like humans. However, this description cannot be considered a definition, as it is highly ambiguous and fuzzy. In addition, the term IA has been evolving based on the knowledge of the moment since it is a very changing science.

Simplifying the term to the current knowledge, we could define **AI as the scientific discipline that deals with creating computer programs that perform operations comparable to those performed by the human mind, such as learning or logical reasoning.**

This tecnology is helping in processes such as healthcare, ecommerce, engineering, finance...

### **Big Data**

Big Data is a term that describes the large volume of data, both structured and unstructured, that floods into business every day. But it is not the amount of data that is important. What matters with Big Data is what organizations do with the data. It can be analyzed for ideas that lead to better strategic business decisions and moves.

It involves datasets or combinations of datasets whose size, complexity and growth rate make it difficult to capture, manage, process or analyze using conventional technologies and tools, such as conventional relational databases and statistics or visualization packages, within the time necessary to make them useful.

Collecting large amounts of data and looking for trends within the data allows to move much more quickly, smoothly and efficiently. It is very useful because it helps organizations to take advange of data and use it to identify new opportunities. That leads to smarter business moves, more efficient operations and higher profits. It can help in:

- **Cost reduction:** identifying more efficient ways of doing business.
- **Faster and better decision making:** analyzing new data and making decisions based on what they have learned.
- **Customization of services:** measuring customer needs and satisfaction through analysis comes to customization of given services.


For example, in healthcare Big Data is very useful in order to analyze large amounts of data such as patient records, health plans, insurance information, etc. and without Big Data it is very difficult to manage. Analyzing these large amounts of data can lead to better and quickly diagnoses and treatments.
