# Correlation: Inner workings of data

After analyzing the distribution, we can also check the correlation. If we pay attention to the correlation between our feature to predict **cardio** and the features **age, weight, blood pressure, cholesterol and gluc** extracted before, we can visualize a strong correlation. Hence, they are the features that will have the most weight for AI prediction.

```
#@title Intractive correlation
@interact
def correlation(feature1=list(cardio.select_dtypes('number').columns),
               feature2=list(cardio.select_dtypes('number').columns)):
    print(f"Correlation: {cardio[feature1].corr(cardio[feature2])}")
```

```
#@title Plot features correlation
correlation_data = cardio.corr()
sns.heatmap(correlation_data, cmap="PiYG")
```
